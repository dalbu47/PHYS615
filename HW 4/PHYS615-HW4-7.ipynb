### Part A

```python
import numpy as np
import matplotlib.pyplot as plt

k = 10
x = np.linspace(-10,10,10000)
U = k*x**4
plot = plt.plot(x, U, color='cyan', label="Potential Energy of Object (k=10)")
plt.xlabel("Distance (x)")
plt.ylabel("Potential Energy (U)")
plt.title("Potential Energy vs Distance")
plt.legend()
```

If given a sharp kick to the right at t = 0, the object will move to the right at a high velocity initially, but will quickly slow down and come back to the minimum at x = 0. The object will then oscillate back and forth from the maximum on one side to the other.

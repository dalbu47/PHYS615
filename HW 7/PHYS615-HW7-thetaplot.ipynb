```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

```python
g = 9.81 # m/s^2
L = 1 # m

def phi_derivatives(x,t):
    return [x[1], (-g/L)*np.sin(x[0])]

time = np.arange(0, 2*np.pi, 1e-3)


ilist = [-np.pi, 0, np.pi]
colorlist = ['red', 'cyan', 'magenta']
label=['IC: = (pi/6,-pi)', 'IC: = (pi/6,0)', 'IC: = (pi/6,pi)']
for i,c,l in zip(ilist,colorlist,label):
    phi, dphi = odeint(phi_derivatives, [np.pi/6, i], time).T
    plt.plot(time, phi, f'{c}', label=f'{l}')
    plt.xlabel('time (s)')
    plt.ylabel('angle (Rad)')
    plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
    plt.axis((0,2*np.pi,-1*np.pi,np.pi))
    plt.ylim(-6,6)
    plt.grid(True)
    plt.legend()
    plt.title("Angle vs Time")
```

```python
plt.plot(time, dphi*10, 'magenta', linewidth = 2)
plt.xlabel('time (s)')
plt.ylabel('angular velocity (rad/s)')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.grid(True)
plt.axis((0, 2*np.pi, -30, 30));
```

```python
_, a = phi_derivatives([phi, dphi], time)
plt.plot(time, a, 'cyan', linewidth = 2)
plt.xlabel('time (s)')
plt.ylabel('angular acceleration (rad/s^2)')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.grid(True)
plt.axis((0, 2*np.pi, -10, 10));
```

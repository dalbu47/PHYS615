```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
```

#### $\large{\text{Bead on a circular wire}}$
$\text{Solving: } \ddot{\phi} = (\omega^2\cos{\phi} - \frac{g}{R})\sin{\phi}\text{ , where }\omega^2>\frac{g}{R}\text{ , }R=1\text{\hspace{1mm}m, and }g = 9.81 \frac{\text{m}}{\text{s}^2}$

```python
g = 9.81 # m/s^2
R = 1 # m
w = 4 # s^-1

def phi_derivatives(x,t):
    return [x[1], (((w**2)*(np.cos(x[0]))) - (g/R))*np.sin(x[0])]

time = np.arange(0, 2*np.pi, 1e-3)
phi, dphi = odeint(phi_derivatives, [np.pi/2, 0], time).T   
```

#### $\text{Initial Conditions}$
$\phi_o = \frac{\pi}{2}\text{ , } \dot{\phi_o}=0$

```python
plt.plot(time, phi, 'red', linewidth = 2)
plt.xlabel('time (s)')
plt.ylabel('Angle (Rad)')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.axis((0,2*np.pi,-1*np.pi,np.pi))
plt.grid(True)
```

```python
plt.plot(time, dphi*10, 'magenta', linewidth = 2)
plt.xlabel('time (s)')
plt.ylabel('angular velocity (rad/s)')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.grid(True)
plt.axis((0, 2*np.pi, -30, 30));
```

```python
_, a = phi_derivatives([phi, dphi], time)
plt.plot(time, a, 'cyan', linewidth = 2)
plt.xlabel('time (s)')
plt.ylabel('angular acceleration (rad/s^2)')
plt.ticklabel_format(style='sci', axis='x', scilimits=(0,0))
plt.grid(True)
plt.axis((0, 2*np.pi, -10, 10));
```
